```python
#PROBLEM 1 — Smart Parking Lot System
def parking_system(capacity, logs):
    parked = 0
    peak = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        peak = max(peak, parked)

    status = "Available"
    if parked > capacity:
        status = "Exceeded Capacity"

    print("Currently Parked Vehicles:", parked)
    print("Parking Status:", status)
    print("Peak Usage:", peak)


# Sample Input
parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
#PROBLEM 2 — Food Delivery Time Estimator
def delivery_time(distance, traffic, weather):
    time = distance * 5   # base 5 mins per km

    if traffic.lower() == "high":
        time += 10
    elif traffic.lower() == "medium":
        time += 5

    if weather.lower() == "rainy":
        time += 15
    elif weather.lower() == "storm":
        time += 25

    print("Estimated Delivery Time:", time, "minutes")


# Sample Input
delivery_time(8, "High", "Rainy")
```

```python
#PROBLEM 3 — Movie Theatre Occupancy Analyzer
def theatre_occupancy(total_seats, booked_seats):
    booked = len(booked_seats)
    occupancy = (booked / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 70:
        status = "Almost Full"
    else:
        status = "Available"

    print("Occupancy:", round(occupancy), "%")
    print("Show Status:", status)


# Sample Input
theatre_occupancy(200, [1]*150)
```

```python
#PROBLEM 4 — Cloud Server Load Classification
def server_load(cpu_readings):
    avg = sum(cpu_readings) / len(cpu_readings)

    if avg < 50:
        status = "Normal"
    elif avg <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", round(avg), "%")
    print("Server Status:", status)


# Sample Input
server_load([45, 60, 70, 85, 90])
```

```python
#PROBLEM 5 — Smart Classroom Resource Monitor
def resource_monitor(resources):
    overused = []

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    alert = "Yes" if overused else "No"

    print("Overused Resources:", ", ".join(overused))
    print("Energy Alert:", alert)


# Sample Input
resource_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

```python
#PROBLEM 6 — Event Registration Controller
def event_registration(capacity, registrations):
    confirmed = min(capacity, registrations)
    waitlisted = max(0, registrations - capacity)

    status = "Closed" if registrations >= capacity else "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Sample Input
event_registration(100, 105)
```
